```python
##Question 1: Extract Tesla Stock Data Using yfinance
```

```python
!pip install yfinance
import yfinance as yf
import pandas as pd

tesla_data = yf.download('TSLA', start='2010-01-01', end='2023-01-01')

tesla_data.reset_index(inplace=True)

print(tesla_data.head())
```

```python
##Question 2: Webscraping to Extract Tesla Revenue Data
```

```python
!pip install beautifulsoup4 requests
import requests
from bs4 import BeautifulSoup

url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')

tables = soup.find_all('table')
for table in tables:
    if 'Tesla Quarterly Revenue' in str(table):
        tesla_revenue = pd.read_html(str(table))[0]
        break

tesla_revenue.dropna(inplace=True)
print(tesla_revenue.tail())
```

```python
gme_data = yf.download('GME', start='2010-01-01', end='2023-01-01')

gme_data.reset_index(inplace=True)

print(gme_data.head())
```

```python
##Question 4: Webscraping to Extract GameStop Revenue Data
```

```python
url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')

tables = soup.find_all('table')
for table in tables:
    if 'GameStop Quarterly Revenue' in str(table):
        gme_revenue = pd.read_html(str(table))[0]
        break

gme_revenue.dropna(inplace=True)
print(gme_revenue.tail())
```

```python
##Question 5: Plot Tesla Stock Graph
```

```python
import matplotlib.pyplot as plt

def make_graph(stock_data, title):
    plt.figure(figsize=(14,7))
    plt.plot(stock_data['Date'], stock_data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.show()

make_graph(tesla_data, 'Tesla Stock Price')
```
